# Entropy
The definition of the entropy is the surprise, (1 over the possibility). The higher the possiblity, the lesser surprise.

If we use the term **log(1/possiblity)=-log(possiblity)**, we will find that when it surely will happen, the posibility equals 1 and the entropy equals 0, means, no surprise. 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#why we use log(1/possibility) instead of (1/possibility) to represent the surprise?
fig, ax = plt.subplots()  
x = np.linspace(0,1,100) # 创建x的取值范围
a=1/x
b=np.log(1/x)
c=np.log(x)
ax.plot(x, a, label='1/possibility') 
ax.plot(x, b, label='log(1/possibility)') 
ax.legend()
plt.show()
fig, ax = plt.subplots() 
x = np.linspace(0.8,1,100) # 创建x的取值范围
a=1/x
b=np.log(1/x)
c=np.log(x)
ax.plot(x, a, label='1/possibility') 
ax.plot(x, b, label='log(1/possibility)') 
ax.legend()
plt.show()
```

```python
#the formula of Entropy= sum(p*log(1/p))
fig, ax = plt.subplots() 
x = np.linspace(0,1,100) # range of axis-X
a=-np.log(x)*x
b=-np.log(1-x)*(1-x)
c=a+b
d=-np.log(x)
e=-np.log(1-x)
ax.plot(x, a, label='p-surprise-weighted') 
ax.plot(x, b, label='(1-p)-surprise-weighted') 
ax.plot(x, c, label='sum-surprise-weighted') 
ax.plot(x, d, label='p-surprise') 
ax.plot(x, e, label='(1-p)-surprise') 

ax.legend()
plt.show()
```
